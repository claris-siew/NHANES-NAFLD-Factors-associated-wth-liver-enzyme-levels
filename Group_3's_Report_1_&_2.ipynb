# 3. Background & Rationale

**3.1 Briefly explain the health issue or knowledge gap being addressed.**

Non-alcoholic fatty liver disease (NAFLD), also known as metabolic dysfunction-associated steatotic liver disease (MASLD), is a condition in which excess fat accumulates in the liver of individuals with minimal or no alcohol intake. NAFLD spans a spectrum from simple hepatic steatosis to non-alcoholic steatohepatitis (NASH), which involves inflammation and hepatocellular injury. If left unmanaged, NASH can progress to liver fibrosis, cirrhosis, and even hepatocellular carcinoma, similar to liver damage seen in long-term alcohol use (Mayo Clinic, 2024).

NAFLD often presents without symptoms in its early stages, making it difficult to detect without routine blood tests. Liver enzymes such as alanine transaminase (ALT), aspartate transaminase (AST), and alkaline phosphatase (ALP) are commonly used clinical and research biomarkers for liver injury. In many individuals, asymptomatic elevations in these enzymes may be the first sign of NAFLD. A significant proportion of patients with unexplained ALT or AST elevation are ultimately found to have NAFLD, which accounts for approximately 25–51% of such cases (Oh et al., 2017).

**3.2 Why is this research question important in public health or biomedical science?**

NAFLD has become a leading global health concern and is currently the most common chronic liver disease worldwide, affecting approximately 25% of the global adult population (Younossi et al., 2021). Its prevalence is especially high in Western and Middle Eastern regions, and it is on the rise in Asia, including Singapore. Singapore recently reported a historic increase in NAFLD rates, with approximately one in three adults affected, a trend that poses growing challenges to the nation's healthcare system (Biotech Connection Singapore, 2024).

Beyond liver-specific complications, NAFLD has been associated with elevated risks of type 2 diabetes, cardiovascular disease, and liver-related cancers, thereby amplifying its impact on both individual and population-level health outcomes (Gato et al., 2024). Despite its clinical significance, there are currently no approved medications that can cure NAFLD. While symptoms can be managed, liver transplantation may be the only option in advanced cases. This highlights the urgency for early detection and prevention.

Liver enzymes (ALT, AST, and ALP) remain valuable non-invasive biomarkers for assessing liver stress or injury. Investigating the upstream factors that contribute to their elevation, including hypertension status, environmental toxicants, vitamin E intake, and demographic or lifestyle factors, could improve early risk identification and inform more effective public health and clinical strategies.

# 4. Research question & Objectives

**4.1 Our original research question.**

Among U.S. adults aged 20 and above from NHANES 2007–2018, how does hypertension status, blood mercury levels, dietary intakes of vitamin E, along with age, gender, physical activity, and BMI, relate to levels of liver enzymes ALT, AST, and ALP which are the biomarkers linked to Non-Alcoholic Fatty Liver Disease (NAFLD)?

**4.2 Justification of its novelty with a short literature review.**

Mercury is a widespread environmental toxicant known to induce liver damage, with multiple studies linking blood mercury concentrations to elevated liver enzymes such as ALT and AST (Cave et al., 2010). However, relatively few studies have examined how co-exposures, such as dietary antioxidant intake, may modify the hepatotoxic effects of mercury.

Vitamin E, a lipid-soluble antioxidant, has been proposed to reduce liver inflammation and oxidative stress in patients with non-alcoholic fatty liver disease (NAFLD). A recent review highlights its therapeutic potential in improving liver enzyme profiles and reducing hepatic steatosis (Abera et al., 2024). Similarly, hypertension, traditionally regarded as a cardiovascular issue, has been increasingly linked to NAFLD through mechanisms such as insulin resistance, oxidative stress, and endothelial dysfunction, as demonstrated in a large cross-sectional analysis (Liang et al., 2022).

Demographic and lifestyle factors also play a critical role. Elevated BMI is a known contributor to liver steatosis and enzyme abnormalities (Wang et al., 2024). Gender-based differences exist, where males tend to exhibit higher ALT levels than females, potentially due to hormonal and metabolic differences (Ruhl & Everhart, 2003). Advancing age is associated with cumulative metabolic stress that contributes to elevated liver enzymes (Bertolotti et al., 2014), while physical activity offers a protective effect, even in the absence of weight loss (Keating et al., 2012).

Although prior studies have assessed these associations individually, few have analysed their combined effects or included interaction terms. Research exploring whether antioxidant vitamins may attenuate mercury’s impact on liver function remains scarce. Additionally, much of the literature is based on small or non-representative samples, limiting generalisability. Our study addresses these gaps using a multidimensional, population-level analysis to improve our understanding of modifiable contributors to liver enzyme elevation and NAFLD risk.

**4.3 How our project builds upon or fills gaps in existing studies.**

Existing studies often examine environmental exposures, dietary intake, or lifestyle factors in isolation, overlooking how these variables may interact to affect liver health. Additionally, key modifying factors such as hypertension or physical activity are frequently omitted from statistical models assessing liver enzyme levels.

This study adopts an integrative approach that examines both independent and interactive associations across a wide range of exposures. By leveraging six NHANES cycles (2007–2018), it improves upon the limitations of previous studies by using a large, representative sample to assess relationships with greater precision and population relevance.

This multidimensional design allows for a more nuanced understanding of how toxicants, nutrients, health status, and lifestyle collectively influence liver function. In doing so, the study provides a more complete picture of the modifiable factors that contribute to elevated liver enzyme levels and early indicators of NAFLD, laying the foundation for more targeted public health interventions.

```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
from scipy.stats import shapiro #For the Shapiro-Wilk test
```

```python
# -----------------------------
# Define NHANES years and cycles (2007–2018)
# -----------------------------
years = [ "2007", "2009", "2011", "2013", "2015", "2017"]
cycles = [ "E", "F", "G", "H", "I", "J"]
```

```python
#Next import the demographics data
# -----------------------------
# Step 1: Import and filter demographics - Will import both age and gender
# -----------------------------
demo_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/DEMO_{c}.xpt"
             for y, c in zip(years, cycles)]
df_demo = pd.concat([pd.read_sas(url) for url in demo_urls])
df_demo = df_demo[["SEQN", "RIAGENDR", "RIDAGEYR"]]
```

```python
#Import the Standard biochemical test data

biochem_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/BIOPRO_{c}.xpt"
                for y, c in zip(years, cycles)]
df_biochem = pd.concat([pd.read_sas(url) for url in biochem_urls])
df_biochem = df_biochem[["SEQN", "LBXSATSI","LBXSASSI","LBXSAPSI"]]
```

```python
#blood mercury levels
bloodmercury_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/PBCD_{c}.xpt"
                for y, c in zip(years, cycles)]
df_bloodmercury = pd.concat([pd.read_sas(url) for url in bloodmercury_urls])
df_bloodmercury = df_bloodmercury[["SEQN", "LBXTHG"]]
```

```python
# List of URLs for Day 1 and Day 2
day1_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/DR1TOT_{c}.xpt"
             for y, c in zip(years, cycles)]
day2_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/DR2TOT_{c}.xpt"
             for y, c in zip(years, cycles)]

# Load and concatenate all years
df_day1 = pd.concat([pd.read_sas(url) for url in day1_urls], ignore_index=True)
df_day2 = pd.concat([pd.read_sas(url) for url in day2_urls], ignore_index=True)

# Select relevant columns
df_day1 = df_day1[["SEQN",  "DR1TATOC"]]
df_day2 = df_day2[["SEQN", "DR2TATOC"]]
# Merge on SEQN
df_dietary = pd.merge(df_day1, df_day2, on="SEQN", how="outer")
# Compute average of each vitamin
df_dietary["TATOC_avg"] = df_dietary[["DR1TATOC", "DR2TATOC"]].mean(axis=1)

# Optional: Keep only SEQN and averaged columns
df_dietary_avg = df_dietary[["SEQN", "TATOC_avg"]]
```

```python
#physical activity
physical_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/PAQ_{c}.xpt"
                for y, c in zip(years, cycles)]
df_physical = pd.concat([pd.read_sas(url) for url in physical_urls])
df_physical = df_physical[["SEQN", "PAD615"]]
```

```python
#bmi score (Body measures data)
body_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/BMX_{c}.xpt"
             for y, c in zip(years, cycles)]
df_body = pd.concat([pd.read_sas(url) for url in body_urls])
df_body = df_body[["SEQN", "BMXBMI"]]
```

```python
import pandas as pd
import requests
import io

# NHANES data files
nhanes_files = {
    "2007-2008": {
        "day1": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2007/DataFiles/DR1TOT_E.xpt",
        "day2": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2007/DataFiles/DR2TOT_E.xpt",
        "demo": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2007/DataFiles/DEMO_E.xpt"
    },
    # [Repeat for other cycles as in your original code...]
    "2017-2018": {
        "day1": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2017/DataFiles/DR1TOT_J.xpt",
        "day2": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2017/DataFiles/DR2TOT_J.xpt",
        "demo": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2017/DataFiles/DEMO_J.xpt"
    }
}

# Store results
median_results = []

for cycle, files in nhanes_files.items():
    # Load datasets
    df_day1 = pd.read_sas(io.BytesIO(requests.get(files["day1"]).content), format='xport')
    df_day2 = pd.read_sas(io.BytesIO(requests.get(files["day2"]).content), format='xport')
    df_demo = pd.read_sas(io.BytesIO(requests.get(files["demo"]).content), format='xport')

    # Merge demographics
    df_day1 = pd.merge(df_day1, df_demo[["SEQN", "RIDAGEYR"]], on="SEQN", how="left")
    df_day2 = pd.merge(df_day2, df_demo[["SEQN", "RIDAGEYR"]], on="SEQN", how="left")

    # Record original counts
    original_day1 = df_day1.shape[0]
    original_day2 = df_day2.shape[0]

    # Filter for age ≥ 20
    df_day1 = df_day1[df_day1["RIDAGEYR"] >= 20]
    df_day2 = df_day2[df_day2["RIDAGEYR"] >= 20]

    # Extract TATOC (Vitamin E intake)
    vit_e_day1 = df_day1["DR1TATOC"] if "DR1TATOC" in df_day1.columns else pd.Series(dtype=float)
    vit_e_day2 = df_day2["DR2TATOC"] if "DR2TATOC" in df_day2.columns else pd.Series(dtype=float)

    # Combine and average
    df_combined = pd.DataFrame({
        "vit_e_day1": vit_e_day1.reset_index(drop=True),
        "vit_e_day2": vit_e_day2.reset_index(drop=True)
    })
    df_combined["TATOC_avg"] = df_combined.mean(axis=1, skipna=True)

    # Count non-missing
    retained = df_combined["TATOC_avg"].dropna().shape[0]
    total = max(original_day1, original_day2)  # conservative estimate

    # Calculate median
    median_val = round(df_combined["TATOC_avg"].dropna().median(), 2)

    # Calculate % retained
    percent_retained = round((retained / total) * 100, 2) if total > 0 else 0

    # Store result
    median_results.append({
        "Cycle": cycle,
        "Median Vitamin E (TATOC_avg, Age ≥ 20)": median_val,
        "Original Sample Size": total,
        "After Dropna": retained,
        "% Retained": percent_retained
    })

# Final output
medians_df = pd.DataFrame(median_results)
print(medians_df)
```

```python
import pandas as pd

# URLs for BPQ and DEMO files by cycle
nhanes_urls = {
    "2007-2008": {
        "bpq": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2007/DataFiles/BPQ_E.xpt",
        "demo": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2007/DataFiles/DEMO_E.xpt"
    },
    "2009-2010": {
        "bpq": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2009/DataFiles/BPQ_F.xpt",
        "demo": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2009/DataFiles/DEMO_F.xpt"
    },
    "2011-2012": {
        "bpq": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2011/DataFiles/BPQ_G.xpt",
        "demo": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2011/DataFiles/DEMO_G.xpt"
    },
    "2013-2014": {
        "bpq": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2013/DataFiles/BPQ_H.xpt",
        "demo": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2013/DataFiles/DEMO_H.xpt"
    },
    "2015-2016": {
        "bpq": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2015/DataFiles/BPQ_I.xpt",
        "demo": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2015/DataFiles/DEMO_I.xpt"
    },
    "2017-2018": {
        "bpq": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2017/DataFiles/BPQ_J.xpt",
        "demo": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2017/DataFiles/DEMO_J.xpt"
    }
}

def analyze_hypertension():
    results = []

    for cycle, urls in nhanes_urls.items():
        print(f"Processing {cycle}...")

        try:
            bpq_df = pd.read_sas(urls["bpq"], format='xport', encoding='latin1')
            demo_df = pd.read_sas(urls["demo"], format='xport', encoding='latin1')
        except Exception as e:
            print(f"Error loading data for {cycle}: {e}")
            continue

        # Merge on SEQN
        merged_df = pd.merge(bpq_df, demo_df, on="SEQN", how="inner")

        # Filter for age >= 20
        merged_df = merged_df[merged_df["RIDAGEYR"] >= 20]

        # Hypertension variable: BPQ020 (Have you ever been told by a doctor you had hypertension?)
        # Exclude missing values (NaN or 7, 9 which are Refused/Don't know codes)
        if "BPQ020" not in merged_df.columns:
            print(f"Hypertension variable BPQ020 missing in {cycle}. Skipping.")
            continue

        # According to NHANES codebook, valid values are typically 1=Yes, 2=No; 7=Refused, 9=Don't know
        valid_responses = merged_df["BPQ020"].isin([1, 2])
        filtered = merged_df[valid_responses]

        if filtered.empty:
            print(f"No valid hypertension responses in {cycle}. Skipping.")
            continue

        median_response = filtered["BPQ020"].median()
        sample_size = len(filtered)

        results.append({
            "Cycle": cycle,
            "Median Hypertension Response (Age ≥ 20)": median_response,
            "Sample Size": sample_size
        })

    return pd.DataFrame(results)

# Run and print results
df_hypertension = analyze_hypertension()
print(df_hypertension)

```

```python
import pandas as pd

# Assume `df` contains individual-level NHANES data with:
# LBDLDL = LDL (mg/dL), LBXVIDMS = Vitamin D (ng/mL), Cycle

# Define thresholds
ldl_threshold = 130   # High LDL
vitd_threshold = 20   # Vitamin D deficiency

# Function to calculate % high-risk by cycle
def calc_high_risk_percent(df):
    results = []

    for cycle, group in df.groupby("Cycle"):
        result = {"Cycle": cycle, "N": len(group)}

        # LDL ≥ 130 mg/dL
        if "LBDLDL" in group.columns:
            result["% High LDL (≥130 mg/dL)"] = round((group["LBDLDL"] >= ldl_threshold).mean() * 100, 2)

        # Vitamin D < 20 ng/mL
        if "LBXVIDMS" in group.columns:
            result["% Vit D Deficient (<20 ng/mL)"] = round((group["LBXVIDMS"] < vitd_threshold).mean() * 100, 2)

        results.append(result)

    return pd.DataFrame(results)

# Example usage
# df = pd.read_csv("merged_ldl_vitd_data.csv")  # Your dataset
# high_risk_df = calc_high_risk_percent(df)
# print(high_risk_df)
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pandas as pd
from scipy.stats import kruskal
import warnings

# Suppress potential warnings from statsmodels or pandas
warnings.filterwarnings("ignore")

# List of liver enzymes
enzyme_vars = ['LBXSATSI', 'LBXSASSI', 'LBXSAPSI']

# List of continuous predictors
continuous_predictors = [
    'LBXTHG',       # blood mercury
    'TATOC_avg',    # vitamin E
    'RIDAGEYR',     # age
    'BMXBMI',       # BMI
    'PAD615'        # physical activity
]

print("--- Tests for Continuous Predictors ---")

# Kruskal-Wallis tests (comparing quartiles)
# Note: Re-binning to ensure consistency if adults_df_clean was modified
print("\n### Kruskal-Wallis tests (comparing quartiles) ###")
for var in continuous_predictors:
    binned_col_name = f'{var}_quartiles'
    # Binning safely
    if adults_df_clean[var].dropna().nunique() >= 4:
        adults_df_clean[binned_col_name] = pd.qcut(adults_df_clean[var], 4, labels=['Q1', 'Q2', 'Q3', 'Q4'], duplicates='drop')
    else:
        # Handle cases with fewer than 4 unique values, potentially creating fewer bins
        # Check if there are enough unique values to create at least 2 bins
        if adults_df_clean[var].dropna().nunique() > 1:
             adults_df_clean[binned_col_name] = pd.cut(adults_df_clean[var], bins=adults_df_clean[var].dropna().nunique(), labels=[f'Bin_{i+1}' for i in range(adults_df_clean[var].dropna().nunique())], duplicates='drop')
             print(f"Warning: Not enough unique values in {var} to create distinct quartiles using qcut. Using cut with {adults_df_clean[var].dropna().nunique()} bins instead.")
        else:
            adults_df_clean[binned_col_name] = None # Cannot bin if 1 or fewer unique values
            if adults_df_clean[var].dropna().nunique() == 1:
                 print(f"Warning: Only one unique value in {var}. Cannot create bins for Kruskal-Wallis.")
            else:
                 print(f"Warning: No non-null values in {var}. Cannot create bins for Kruskal-Wallis.")


    for enzyme in enzyme_vars:
        df_temp = adults_df_clean[[binned_col_name, enzyme]].dropna()
        if binned_col_name in df_temp.columns: # Check if binning was successful
            groups = [df_temp[df_temp[binned_col_name] == level][enzyme].dropna() for level in df_temp[binned_col_name].unique()]
            # Filter out empty groups
            groups = [group for group in groups if len(group) > 0]

            if len(groups) >= 2: # Kruskal-Wallis requires at least 2 groups with data
                try:
                    stat, p_val = kruskal(*groups)
                    print(f"Kruskal-Wallis for {enzyme} by {var} quartiles: H={stat:.3f}, p={p_val:.4f}")
                except ValueError as e:
                    print(f"Could not run Kruskal-Wallis for {enzyme} by {var} quartiles: {e}")
            else:
                print(f"Skipping Kruskal-Wallis for {enzyme} by {var} quartiles: Not enough groups with data.")
        else:
             print(f"Skipping Kruskal-Wallis for {enzyme} by {var}: Binning failed or no data.")


print("\n--- End of Tests for Continuous Predictors ---")
```

```python
import pandas as pd
from scipy.stats import ttest_ind
import warnings

# Suppress potential warnings
warnings.filterwarnings("ignore")

# List of liver enzymes
enzyme_vars = ['LBXSATSI', 'LBXSASSI', 'LBXSAPSI']

# List of continuous predictors
continuous_predictors = [
    'LBXTHG',       # blood mercury
    'TATOC_avg',    # vitamin E
    'RIDAGEYR',     # age
    'BMXBMI',       # BMI
    'PAD615'        # physical activity
]

print("--- Independent Samples T-tests for Continuous Predictors (Median Split) ---")

for predictor in continuous_predictors:
    print(f"\n### Testing {predictor} ###")
    # Calculate the median, excluding NaNs
    median_val = adults_df_clean[predictor].median()

    # Create two groups based on the median
    # Using .loc to avoid SettingWithCopyWarning
    adults_df_clean.loc[:, f'{predictor}_group'] = adults_df_clean[predictor].apply(lambda x: 'High' if x >= median_val else 'Low' if pd.notna(x) else x)

    for enzyme in enzyme_vars:
        df_temp = adults_df_clean[[f'{predictor}_group', enzyme]].dropna()

        # Separate data into two groups
        group1 = df_temp[df_temp[f'{predictor}_group'] == 'Low'][enzyme].dropna()
        group2 = df_temp[df_temp[f'{predictor}_group'] == 'High'][enzyme].dropna()

        # Perform independent samples t-test if both groups have data
        if len(group1) > 1 and len(group2) > 1: # t-test requires at least 2 samples in each group
            try:
                stat, p_val = ttest_ind(group1, group2, nan_policy='omit')
                print(f"T-test for {enzyme} by {predictor} (Median Split): t={stat:.3f}, p={p_val:.4f}")
            except ValueError as e:
                print(f"Could not run T-test for {enzyme} by {predictor} (Median Split): {e}")
        else:
            print(f"Skipping T-test for {enzyme} by {predictor} (Median Split): Not enough data in both groups.")

print("\n--- End of T-tests for Continuous Predictors ---")
print("\nNote: Splitting continuous variables into groups for t-tests or ANOVA can lead to loss of information. Linear regression is generally preferred for analyzing the relationship between continuous variables.")
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import warnings

warnings.filterwarnings("ignore")

# Dependent variables (enzymes)
enzyme_vars = ['LBXSATSI', 'LBXSASSI', 'LBXSAPSI']

# Ensure readable categorical labels
if 'Hypertension' not in adults_df_clean.columns:
    adults_df_clean['Hypertension'] = adults_df_clean['BPQ020'].map({1: 'Yes', 2: 'No', 7: 'Refused', 9: "Don't Know"})
if 'Gender' not in adults_df_clean.columns:
    adults_df_clean['Gender'] = adults_df_clean['RIAGENDR'].map({1: 'Male', 2: 'Female'})

# Keep only clean, binary categories for modeling
df = adults_df_clean.copy()
df = df[df['Hypertension'].isin(['Yes', 'No'])]
df = df[df['Gender'].isin(['Male', 'Female'])]

print("--- Multiple Linear Regression (categoricals → binary dummies) ---\n")

for enzyme in enzyme_vars:
    # Build modeling frame
    df_subset = df[[enzyme, 'Gender', 'Hypertension']].copy()

    # Make sure the dependent variable is numeric (coerce weird strings like 'Below LOD' to NaN)
    df_subset[enzyme] = pd.to_numeric(df_subset[enzyme], errors='coerce')

    # Drop rows with missing outcome or predictors
    df_subset = df_subset.dropna(subset=[enzyme, 'Gender', 'Hypertension'])

    # One-hot encode to binary dummies; drop_first avoids dummy trap
    X = pd.get_dummies(df_subset[['Gender', 'Hypertension']], drop_first=True)

    # Coerce all predictors to float (prevents "dtype object" issues)
    X = X.astype(float)

    # Add intercept
    X = sm.add_constant(X)

    # Outcome
    y = df_subset[enzyme].astype(float)

    # Guard against empty data after cleaning
    if len(df_subset) < 2 or X.shape[1] >= len(df_subset):
        print(f"⚠️ Skipping {enzyme}: not enough clean rows after filtering/encoding.")
        continue

    try:
        model = sm.OLS(y, X).fit()
        print(f"📊 Regression for {enzyme}:")
        # Compact table of coefficients and p-values
        res = pd.DataFrame({
            'coef': model.params,
            'std_err': model.bse,
            't': model.tvalues,
            'p>|t|': model.pvalues
        })
        print(res.round(4))
        print(f"\nR² = {model.rsquared:.4f}, Adj. R² = {model.rsquared_adj:.4f}, n = {len(df_subset)}")
        print("\n" + "-"*70 + "\n")
    except Exception as e:
        print(f"❌ Error for {enzyme}: {e}")

print("\n--- End of Multiple Linear Regression ---")
```

Mercury & Vitamin E Interaction: High mercury and high vitamin E intake resulted in significantly different ALT & AST levels compared to most other groups. For ALP, differences were mainly driven by mercury levels, with high mercury groups differing significantly from low mercury groups regardless of Vitamin E intake. This suggests that mercury exposure strongly influences ALP, while the combined effect of mercury and vitamin E impacts ALT and AST. However, this interaction effect was observed only in group comparisons and was not tested using a regression interaction term.

**Public health implications**

From a public health perspective, these findings are relevant for both prevention and screening strategies for NAFLD.  

Environmental Health Action: Mercury exposure remains a clear threat to liver health, justifying stronger environmental monitoring, fish consumption guidelines, and occupational protections.

Targeted Nutritional Strategies: The inconsistent relationship between vitamin E and liver enzymes suggests supplementation policies could be explored in populations with high mercury exposure, though evidence for benefit is not conclusive.

Age-Sensitive Screening: Age-related shifts in liver enzyme patterns indicate that diagnostic thresholds may need adjustment for older adults to prevent under- or over-diagnosis.

Obesity Prevention and Early Liver Screening: Strong associations between BMI and enzyme changes reinforce the need for obesity reduction initiatives and routine liver health checks in overweight populations.

Physical Activity Guidance: Public health campaigns should emphasize sustained, moderate-to-high physical activity levels, as benefits to liver health may not be linear.

Integrated Cardiometabolic and Liver Health Programs: Having hypertension was statistically significantly linked to higher levels of all three enzymes, although the effect sizes were small. These associations support coordinated screening strategies that address cardiovascular and hepatic risk together.

Gender-Sensitive Awareness: Males had higher ALT and AST compared to females, suggesting targeted awareness and screening may be warranted for men.

Complex Toxin–Nutrient Interactions: Interventions should account for possible interactions between environmental toxins and nutrients, recognizing that nutritional supplementation cannot fully offset toxin-related damage.

**Limitations of NHANES data**

While these findings offer valuable insights, there are a few limitations of NHANES data that should be considered when interpreting the results. NHANES is a cross-sectional survey, which limits causal inference and only shows associations at a single point in time. This limitation particularly affects the interpretation of interactions (e.g., mercury × vitamin E), which cannot be confirmed as causal.

Self-reported data such as physical activity can be affected by memory errors or incorrect reporting. Additionally, biomarker measurements like liver enzymes and mercury levels mainly reflect short-term or recent exposures rather than long-term patterns, which may underestimate or overestimate chronic effects. Although the survey is nationally representative, rare exposures or subpopulation effects may not be well-reflected in the data. Finally, residual confounding from unmeasured factors, such as other environmental toxins, genetic predispositions, or detailed dietary patterns, may influence observed associations.

9.2 Future work

**Further analysis**

While the current analysis provides important cross-sectional associations between mercury exposure, vitamin E intake, demographic and lifestyle factors, and liver enzyme levels, additional analytical approaches could yield deeper insights into the determinants of NAFLD risk.

- Logistic Regression for NAFLD Risk Classification

Convert liver enzyme biomarkers into binary outcomes (normal vs. elevated, based on clinical cut-offs) and model the odds of elevation as a function of environmental, dietary, and lifestyle predictors.

This would allow for adjustment of multiple covariates simultaneously, estimation of adjusted odds ratios, and identification of the most influential risk factors for abnormal liver enzyme profiles.

Interaction terms (e.g. BMI × physical activity, hypertension × gender) could be incorporated to explore synergistic or mitigating effects, particularly given that some main effects (e.g., vitamin E, physical activity) were non-significant in regression models but showed differences in group comparisons.

- Temporal Trend Analysis

Extend the descriptive annual trend analysis by applying age- and gender-adjusted regression models to assess changes in enzyme levels and key predictors across NHANES cycles.

This would help determine whether observed shifts (e.g., the recent ALP spike) are statistically significant and whether they persist after adjusting for demographic shifts or survey design effects.

- Stratified Analyses

Repeat analyses within subgroups defined by age, gender, BMI category, or hypertension status to detect effect modification.

This could clarify whether certain risk factors are particularly important in specific populations, for example, whether hypertension effects are stronger in unadjusted comparisons or whether gender modifies the impact of mercury exposure.

- Sensitivity Analyses

Assess the robustness of findings by testing different cut-off values for “high” vs. “low” mercury, vitamin E, and physical activity, as well as by excluding outliers in enzyme levels.
